```python
# Personal Finance Tracker

import os
import hashlib
from datetime import datetime

TRANSACTION_FILE = "transactions.csv"
USER_FILE = "users.txt"

# -------------------------------
# Transaction Class
# -------------------------------
class Transaction:
    def __init__(self, date, category, amount):
        self.date = date
        self.category = category
        self.amount = amount

    def __str__(self):
        return f"{self.date} | {self.category} | ${self.amount:.2f}"

# -------------------------------
# File Handling
# -------------------------------
def load_transactions():
    transactions = []

    if not os.path.exists(TRANSACTION_FILE):
        open(TRANSACTION_FILE, "w").close()  # creates file if missing
        return transactions

    try:
        with open(TRANSACTION_FILE, "r") as file:
            for line in file:
                parts = line.strip().split(",")
                if len(parts) == 3:
                    try:
                        date, category, amount = parts
                        transactions.append(Transaction(date, category, float(amount)))
                    except ValueError:
                        print("Skipping corrupted line:", line.strip())
    except Exception as e:
        print("Error loading file:", e)

    return transactions


def save_transactions(transactions):
    try:
        with open(TRANSACTION_FILE, "w") as file:
            for t in transactions:
                file.write(f"{t.date},{t.category},{t.amount}\n")
    except Exception as e:
        print("Error saving file:", e)

# -------------------------------
# Authentication
# -------------------------------
def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()


def login():
    if not os.path.exists(USER_FILE):
        print("No user found. Create a new account.")
        username = input("Create a username: ")
        password = input("Create a password: ")

        with open(USER_FILE, "w") as file:
            file.write(f"{username},{hash_password(password)}\n")

        print("Account created! Please login again.\n")
        return False

    attempts = 3

    while attempts > 0:
        username = input("Username: ")
        password = input("Password: ")
        hashed = hash_password(password)

        try:
            with open(USER_FILE, "r") as file:
                for line in file:
                    stored_user, stored_pass = line.strip().split(",")
                    if username == stored_user and hashed == stored_pass:
                        print("Login successful!")
                        return True
        except Exception as e:
            print("Error reading user file:", e)

        attempts -= 1
        print(f"Invalid credentials. Attempts left: {attempts}")

    print("Too many failed attempts. Access denied.")
    return False

# -------------------------------
# Transaction Functions
# -------------------------------
def add_transaction(transactions):
    try:
        date = input("Enter date (YYYY-MM-DD): ").strip()

        # Validate date format
        try:
            datetime.strptime(date, "%Y-%m-%d")
        except ValueError:
            print("Invalid date format. Use YYYY-MM-DD.")
            return

        category = input("Enter category: ").strip()
        if not category:
            print("Category cannot be empty.")
            return

        amount = float(input("Enter amount: "))

        if amount == 0:
            print("Amount cannot be zero.")
            return

        transactions.append(Transaction(date, category, amount))
        save_transactions(transactions)
        print("Transaction added successfully.")

    except ValueError:
        print("Invalid input. Amount must be a number.")


def display_transactions(transactions):
    if not transactions:
        print("No transactions found.")
        return

    print("\n--- Transactions ---")
    for i, t in enumerate(transactions, 1):
        print(f"{i}. {t}")


def delete_transaction(transactions):
    display_transactions(transactions)

    if not transactions:
        return

    try:
        index = int(input("Enter transaction number to delete: "))

        if index < 1 or index > len(transactions):
            print("Invalid index. Please select a valid number.")
            return

        removed = transactions.pop(index - 1)
        save_transactions(transactions)
        print("Deleted:", removed)

    except ValueError:
        print("Invalid input. Please enter a number.")

# -------------------------------
# Main Menu
# -------------------------------
def main():
    transactions = load_transactions()

    if not login():
        print("Access denied")
        return

    while True:
        print("\n1. Add Transaction")
        print("2. Display Transactions")
        print("3. Delete Transaction")
        print("4. Exit")

        choice = input("Enter choice: ").strip()

        if choice == "1":
            add_transaction(transactions)
        elif choice == "2":
            display_transactions(transactions)
        elif choice == "3":
            delete_transaction(transactions)
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please select 1-4.")

# -------------------------------
# Run Program
# -------------------------------
if __name__ == "__main__":
    main()
```
